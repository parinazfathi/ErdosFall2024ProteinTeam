```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold
```

```python
df_full = pd.read_csv("/Users/parinazfathi/Documents/GitHub/ErdosFall2024ProteinProfiles/Combined_df2_imputedmean.csv")
proteins=df_full.columns[5:]
```

```python
# Encode the cancer types as numbers

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df_full['Cancer'] = le.fit_transform(df_full['Cancer'])
#df_full.head(15)
```

```python
df_full.head(15)
```

```python
# Split the data into X and y
X = df_full[proteins]
y = df_full['Cancer']
```

```python
X
```

```python
# Split data into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify = y)
```

DECISION TREE CLASSIFIER

```python
from sklearn.feature_selection import RFECV 
from sklearn.tree import DecisionTreeClassifier 
```

```python
# Prepare the decision tree classifier
estimator = DecisionTreeClassifier(random_state=42) 
```

```python
selector = RFECV(estimator,cv = StratifiedKFold(5), step=1) 
selector = selector.fit(X_train, y_train) 
```

```python
# Print the optimal number of features 
print("Optimal number of features: %d" % selector.n_features_) 
```

```python
# Print the selected features 
print("Selected features: %s" % selector.support_) 

print("\nSelected features:\n")
selected_features = X.columns[selector.support_]
print(selected_features)
```
