```python
# Import the packages we may need
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Read in the Gonzalez paper data
Gonzalez_df=pd.read_excel("/Users/parinazfathi/Documents/GitHub/ParinazProteinProfiles/Gonzales_CancerRes_TableS2.xlsx")
```

```python
# Check what it looks like
Gonzalez_df.head(15)
```

```python
# Concatenate the UniprotID column  with the columns for each sample

Gonzalez_df_edited = pd.concat([Gonzalez_df['UniprotID'], Gonzalez_df.iloc[:,4:]], axis=1)
Gonzalez_df_edited.head(15)
```

```python
# Transpose the dataframe and get rid of the number index
Gonzalez_df_edited_transposed = Gonzalez_df_edited.set_index('UniprotID').T
print(Gonzalez_df_edited_transposed)
```

```python
# Read in the list of proteins in common

Proteinsincommon = pd.read_excel("/Users/parinazfathi/Documents/GitHub/ErdosFall2024ProteinProfiles/Proteinsincommon.xlsx")
```

```python
# Take only the columns that overlap with the proteins of interest
Gonzalez_df_shortened = Gonzalez_df_edited_transposed[Proteinsincommon.columns]
Gonzalez_df_shortened.head(15)
```
