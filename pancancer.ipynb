```python
#convert txt file to csv file. Defined as "df1", since it redefines as "df" to ease of use after columns are seperated
#initially, data frame defined as 'df1',  after separating columns for ease of use it redefined as 'df' for all later use
#pancancer.txt dataset downloaded from "https://pmc.ncbi.nlm.nih.gov/articles/PMC10354027/#notes4" 

import pandas as pd
df1 = pd.read_fwf('pancancer.txt')
df1.to_csv('pancancer_old.csv')
df1
```

```python
# since both txt and csv files behave they have only one combined column, it has to be splitted different colums.

df = df1['"Sample_ID" "Cancer" "Assay" "OlinkID" "UniProt" "Panel" "NPX"'].str.split(' ', expand= True)
df
```

```python
#after splitting columns as desired, name them as stated in the file

df.columns=['Sample_ID', 'Cancer', 'Assay', 'OlinkID', 'UniProt', 'Panel', 'NPX']

df
```

```python
# Create a new CSV file from the split DataFrame
output_csv_file = 'pancancer.csv'

# Save the split DataFrame to a CSV file
df.to_csv(output_csv_file, index=False)

print(f"Successfully saved the split DataFrame to '{output_csv_file}'.")
df
```

```python
#number of rows of txt file

def count_rows(file_path):
    with open(file_path, 'r') as file:
        row_count = 0
        for line in file:
            row_count += 1
        return row_count

file_path = 'pancancer.txt'  # Replace 'your_file.txt' with the actual file path
rows = count_rows(file_path)
print(f"The file has {rows} rows.")
```

```python
## Print the last 5 rows of txt file

def read_last_n_lines(file_path, n):
    """Read the last n lines from a text file."""

    with open(file_path, 'r') as f:
        lines = f.readlines()
        return lines[-n:]

file_path = 'pancancer.txt'  # Replace with your file path
n = 5  # Number of lines to read from the end

last_lines = read_last_n_lines(file_path, n)

for line in last_lines:
    print(line, end='')  # The 'end' argument prevents extra newlines
```
